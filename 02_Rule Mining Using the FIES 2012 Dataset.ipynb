# Dataset Description (Empeno)

## Family Income and Expenditure Survey 2012 Volume 1

### Brief Description: 

The Family Income and Expenditure Survey (FIES) is a survey on family income and expenditure of households undertaken by the Philippine Statistics Authority (PSA) across the country. It is the main source of data on family income and expenditure in the Philippines.

The 2012 FIES is a sample survey designed to provide income and expenditure data for the country and its seventeen (17) regions. These data are the basic sources of information on human development and the basis of the policies and programs of the government to improve the socio-economic conditions of the Filipino people.

This volume and edition of the Family Income and Expenditures Survey (FIES) contains data about expenditures and other disbursements, housing characteristics, social protection, income and other receipts and entrepreneural activities of families around the Philippines for the year 2012.

### Collection Process:

The 2012 FIES involved the interview of a national sample of about 50,000 sample households deemed sufficient to gather data on family income and family expenditure and related information affecting income and expenditure levels and patterns in the Philippines at the national and regional level. The sample households covered in the survey were the same households interviewed in the July 2012 and January 2013 round of the LFS.

The PSA carries out nationwide regular household-based surveys such as that of FIES every three years. To be more efficient in the conduct of surveys, PSA designs a master sample consists of randomly assigned and selected set of geographic areas with non-overlapping and discernable boundaries known as the primary sampling units (PSUs). The primary sampling unit (PSU) can be (1) the whole barangay, or (2) a portion of a large barangay, or (3) combinations of small barangays.

The 2012 FIES used the 2003 Master Sample, which relied on a multi-stage area sampling design. First, barangays or combinations of barangays with at least 500 households were grouped into primary sampling units (PSUs). The 17 regions of the Philippines served as the primary strata, and within each region, further divisions were made using geographic and socioeconomic factors like agriculture and income levels. Some PSUs were so large they were automatically included (certainty PSUs), while smaller ones were randomly selected (non-certainty PSUs) based on the number of households from the 2000 census. Once PSUs were chosen, enumeration areas (EAs) within them were selected, followed by housing units in the EAs. In most cases, all households in the sampled housing units were surveyed, except for housing units with more than three households, where a random sample of three households was taken.

The interview involved a questionnaire. The questionnaire collects information on family income, sources of income, family expenditure and related information affecting family income and expenditure levels and patterns in the Philippines.

### Implications of the Data Collection Method:

The 2012 FIES sampling method ensures geographical representativeness by stratifying regions and socioeconomic factors which enables accurate region-specific insights. However, small areas may be underrepresented due to the random selection of non-certainty PSUs which leads to potential sampling bias. Stratification by income and agricultural engagement improves insights on disparities, but reliance on outdated 2000 census data may reduce the relevance of findings for areas that are rapidly changing. The limitation on surveying only three households in larger units could skew results for large households, affecting conclusions on income and family size. Larger regions benefit from more precise estimates, but smaller or less populated areas may lack accuracy. Lastly, the time lag between the 2000 census and 2012 survey introduces the risk that socioeconomic changes in the intervening years could affect the applicability of the findings.

### How does the total household income (**`TotalIncome`**) vary between urban (**`Urban`**) and rural households?

```python
#Group by Urban/Rural classification and calculate descriptive statistics for TotalIncome
urban_income_summary = expenditure_df.groupby('Urban')['TotalIncome'].agg(['mean', 'median', 'std']).reset_index()
print(urban_income_summary)
```

Urban households have a higher mean and median income compared to rural households. The mean income for urban households is 296,253, with a median of 204,615, while rural households have a mean income of 168,870 and a median of 113,148. Additionally, the standard deviation is larger for urban households (321,153) compared to rural households (184,700). This means a larger variability in income among urban households. What this means is that urban areas tend to have both higher incomes and more income inequality than rural areas.

```python
#Values without the outliers can be obtained through the interquartile range. Any data points outside of the range defined by Q1 - 1.5 * IQR and Q3 + 1.5 * IQR are considered outliers.
#Define the function to remove outliers using the IQR method
def remove_outliers(df, column):
    #Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    #Calculate the IQR (Interquartile Range)
    IQR = Q3 - Q1
    
    #Define the range to consider as non-outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    #Filter out the rows that have outliers
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
#Remove outliers from the 'TotalIncome' column
expenditure_df_no_outliers = remove_outliers(expenditure_df, 'TotalIncome')

#Group by Urban/Rural classification and calculate mean, median, and standard deviation
urban_income_summary_no_outliers = expenditure_df_no_outliers.groupby('Urban')['TotalIncome'].agg(['mean', 'median', 'std']).reset_index()

#Print the summary without outliers
print(urban_income_summary_no_outliers)
```

Rural households have an average income of 139,152, with a median of 109,000, indicating a right-skewed distribution where a few high-income households raise the mean. The standard deviation of 96,882.71 indicates a significant variation in income. On the other hand, urban households have a higher average income of 207,576, with a median of 180,615, which is closer to the mean. This suggests a less skewed distribution compared to rural areas. The standard deviation in urban areas is 114,798.56, indicating that while urban areas have higher overall incomes, they also exhibit a larger income spread. This means urban areas have both high-income and low-income households. Overall, urban households earn, on average, about 50% more than rural households.

The significance of this question lies in its potential to help policymakers develop targeted interventions. ### Binning (Pineda)

Binning is necessary since we have numerical variables (`TotalIncome`, `FamilySize`, `TotalExpenditures`, `Age`). This is to bin them into discrete categories and treat these categories as items in the "market basket".

The bins for income and expenditures are split into income classes based on the study of `Albert et al. (2015)` which profiled what income and expenditure ranges in the FIES 2012 are necessary to call a household poor to rich.

These bins are:

 - `Poor:` <94,680
 - `Low Income But Not Poor:` Between 94,680 to 189,360
 - `Lower Middle Income:` Between 189,360 to 378,720
 - `Middle Income:` Between 378,720 to 946,800
 - `Upper Middle Income:` Between 946,800 to 1,420,200
 - `Upper Income But Not Rich:` Between 1,420,200 to 1,893,600
 - `Rich:` >1,893,600
 
Source: Albert, J. , Gaspar, J., & Raymundo, M. (2015). Why we should pay attention to the middle class. `https://pidswebs.pids.gov.ph/CDN/PUBLICATIONS/pidspn1513.pdf`

```python
#Set the bins and bin labels
bins = [0, 94680, 189360, 378720, 946800, 1420200, 1893600, float('inf')]
labels = ['Poor', 
          'Low Income But Not Poor', 
          'Lower Middle Income', 
          'Middle Income', 
          'Upper Middle Income', 
          'Upper Income But Not Rich', 
          'Rich']

#Bin the data
expenditure_df['IncomeBin'] = pd.cut(expenditure_df['TotalIncome'], bins=bins, labels=labels)
expenditure_df['ExpenditureBin'] = pd.cut(expenditure_df['TotalExpenditures'], bins=bins, labels=labels)

#Set the bins and labels for FamilySize
bins = [0, 2, 4, float('inf')]
labels = ['Small (1 - 2)', 'Medium (3 - 4)', 'Large (5+)']

#Apply the bins to FamilySize
expenditure_df['FamilySizeBin'] = pd.cut(expenditure_df['FamilySize'], bins=bins, labels=labels)

#Define the bins and labels for Age
bins = [14, 34, 59, float('inf')]
labels = ['Young (15 - 34)', 'Middle-aged (35 - 59)', 'Senior (60+)']

#Apply the bins to the Age column
expenditure_df['AgeBin'] = pd.cut(expenditure_df['Age'], bins=bins, labels=labels)
```

### Unpaired Observations
Calling back to our exploratory data analysis, we will perform hypothesis testing on those questions to get conclusive statistical evidence to accept or reject our null hypothesis.

We know that income will eventually be binned, but in the case that it was not binned yet, and we want to test between two-group variables like urban status, electricity access, and household head's sex.

**For urban status, the hypotheses are:**

$H_0$: There is no significant difference in household income between urban and rural areas.

$H_A$: There is a significant difference in household income between urban and rural areas.

**For electricity status, the hypotheses are:**

$H_0$: There is no significant difference in household income between households who have electricity and households who don't.

$H_A$: There is a significant difference in household income between households who have electricity and households who don't.

**For household head's sex, the hypotheses are:**

$H_0$: There is no significant difference in household income between male-headed and female-headed household heads.

$H_A$: There is a significant difference in household income between male-headed and female-headed household heads.

```python
#LETS DEAL WITH URBAN STATUS FIRST

import scipy.stats as stats #Import the stats module from scipy for statistical tests

#Separate data for Urban and Rural areas based on the 'Urban' column
urban_data = expenditure_df[expenditure_df['Urban'] == 'Urban']['TotalIncome'] #TotalIncome for Urban areas
rural_data = expenditure_df[expenditure_df['Urban'] == 'Rural']['TotalIncome'] #TotalIncome for Rural areas

#Perform an independent t-test between urban and rural income data, assuming unequal variances
t_stat, p_value = stats.ttest_ind(urban_data, rural_data, equal_var=False)
```

In our code, we're performing an independent t-test (stats.ttest_ind) assuming unequal variances between the urban and rural groups. The t-test, particularly the one used here with the equal_var=False parameter (Welch’s t-test), assumes that the data in each group (urban and rural) follows a normal distribution. Let's test for normality.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

#Visualize the distribution for both urban and rural data
plt.figure(figsize=(12, 6))

#Urban data distribution
plt.subplot(1, 2, 1)
sns.histplot(urban_data, kde=True)
plt.title('Urban TotalIncome Distribution')

#Rural data distribution
plt.subplot(1, 2, 2)
sns.histplot(rural_data, kde=True)
plt.title('Rural TotalIncome Distribution')

plt.tight_layout()
plt.show()

#Perform Shapiro-Wilk test for normality
urban_shapiro = stats.shapiro(urban_data)
rural_shapiro = stats.shapiro(rural_data)

print(f"Shapiro-Wilk test for Urban: Statistic={urban_shapiro[0]}, p-value={urban_shapiro[1]}")
print(f"Shapiro-Wilk test for Rural: Statistic={rural_shapiro[0]}, p-value={rural_shapiro[1]}")
```

The p-value of 0.00 for both is less than the alpha level of 0.05, so this means the data significantly deviates from normality, and the t-test may not be appropriate based on a significance levle of 5%. Let's use a test that does not require the data to be normal. This is the Mann-Whitney U Test.

```python
#We now use a non-parametric test to do statistical inference.
#This is the Mann-Whitney U test which does not need normality for it to be done.
u_stat, u_p_value = stats.mannwhitneyu(urban_data, rural_data)
print(f'Mann-Whitney U test statistic: {u_stat}, p-value: {u_p_value}')
```

```python
#Print the t-statistic and p-value of the test
print(f'T-statistic: {u_stat}')#Display the t-statistic
print(f'P-value: {u_p_value}')#Display the p-value

alpha = 0.05 #SIGNIFICANCE LEVEL OF 5%

#Check if the p-value is less than the alpha value to determine statistical significance
if u_p_value < alpha:
    print("There is a significant difference between urban and rural total income.") #Conclusion if significant
else:
    print("There is no significant difference between urban and rural total income.") #Conclusion if not significant
```

The Mann-Whitney U test results show a test statistic of 269621818.5 and a p-value of 0.0. Since the p-value is less than the alpha level of 0.05, we can conclude that there is a significant difference in total income between urban and rural households based on the Mann-Whitney U test at a significance level of 5%.

 - **`Sex`**    

The hypotheses are clearly stated:

$H_0$: Household income is independent of **[Independent Variable]**  
$H_A$: Household income is dependent on **[Independent Variable]**

```python
from scipy.stats import chi2_contingency

#List of independent variables to test
independent_vars = [
    "Region", "Urban", "Electricity", "FamilySizeBin",
    "EducationLevel", "ExpenditureBin", "AgeBin", "Sex"
]

#Dependent variable
dependent_var = "IncomeBin"

#Function to compute Chi-Square test for each pair
results = []
for var in independent_vars:
    #Create a contingency table
    contingency_table = pd.crosstab(expenditure_df[dependent_var], expenditure_df[var])

    #Perform Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    #Append results
    results.append({
        "Variable": var,
        "Chi2 Statistic": chi2,
        "P-Value": p,
        "Degrees of Freedom": dof
    })

#Display results as a DataFrame
results_df = pd.DataFrame(results)
print(results_df)
```

From the table above, we can observe that
 - The Chi-Square statistic is quite large for all independent variables
 - The P-value for all independent variables is **0.00**

These results suggest that, at a significance level of **`5%`**, since each of these variables' p-values are 0.00 and 0.00 is less than 0.05, we reject the null hypothesis that household income is independent of the described variables. Thus, household income is significantly dependent on each of these independent variables.

Note that a sufficiently large sample size is generally needed to ensure the validity of the test. The larger the sample size, the more reliable the test results will be. Since the FIES 2012 has over 40,000 observations this benefits us in terms of results validity. No need to check for other assumptions for these tests of independence.

Let us instantiate a `RuleMiner` object with relative support threshold `support_t` equal to `0.09` and `confidence_t` equal to `0.7`. The field `support_t` represents the support threshold, while the field `confidence_t` represents the confidence threshold.

A higher relative support would typically indicate a rule that applies to a significant portion of the population. Meanwhile, we want rules that are highly reliable so let's filter rules where the antecedent strongly predicts the consequent. The two have to be balanced because too high of a support and confidence can generate no rules at all.

There is another file inside the folder where this file is located, which is `rule_miner.py`. The `RuleMiner` class is designed to perform association rule mining using a dataset, specifically through the identification of frequent itemsets and the generation of association rules based on those itemsets. The mining process uses support and confidence thresholds to determine significant itemsets and rules. The class provides methods to calculate support for itemsets, generate itemset combinations, and evaluate association rules using predefined thresholds.

It has 7 components:
 - The `__init__` function initializes the RuleMiner object, setting the support and confidence thresholds. It also defines whether support should be calculated as relative (proportion of transactions) or absolute (count of occurrences). This constructor method accepts three arguments: support_t for the support threshold, confidence_t for the confidence threshold, and relative_support to determine if the support should be calculated relative to the total number of transactions or as an absolute count.
 - The `get_support` function calculates the support for a given itemset. It first checks which rows in the dataset contain all the items in the itemset. The support is then calculated either as the absolute count (if relative_support is set to False) or as the proportion of transactions (if relative_support is True). This method is essential for evaluating the frequency of itemsets within the dataset.
 - The `merge_itemsets` function merges smaller itemsets into larger ones. It takes a list of itemsets and combines any pairs of itemsets that share a common item, creating new itemsets that are one item larger. For instance, if there are itemsets like [1, 2], [1, 3], and [2, 6], they are merged to form new itemsets such as [1, 2, 3], [1, 2, 6], and [1, 3, 6]. This process is important for generating larger itemsets that might be more meaningful in rule generation.
 - The `get_rules` function generates all possible association rules from a given itemset. It works by generating all non-empty subsets of the itemset and then pairing each subset as the left-hand side (LHS) of the rule, with the remaining items as the right-hand side (RHS). For example, if the itemset is [1, 2, 3], it will generate rules like [1] -> [2, 3], [2] -> [1, 3], and [1, 2] -> [3]. This method is key for creating the rules that can be evaluated based on their support and confidence.
 - The `get_frequent_itemsets` function identifies all frequent itemsets in the dataset that meet or exceed the specified support threshold. The process starts by considering single-item itemsets, which are the individual columns of the dataset. Then, it iterates through itemsets, checking if their support meets the threshold. If it does, the itemset is added to a list of frequent itemsets. The algorithm continues to expand the itemsets by merging smaller itemsets into larger ones using the merge_itemsets function. This process repeats until no new frequent itemsets can be generated, at which point the list of frequent itemsets is returned.
 - The `get_confidence` function calculates the confidence of an association rule. Given a rule in the form of [lhs, rhs], where lhs is the left-hand side and rhs is the right-hand side, the confidence is computed as the support of the concatenated lhs and rhs itemsets divided by the support of just the lhs itemset. If the support of lhs is zero, meaning it doesn't appear in any transactions, the confidence is returned as zero to avoid division by zero. Otherwise, the confidence value is returned, representing how likely it is that the items in the rhs will appear when the items in lhs appear.
 - The `get_association_rules` function generates and returns a list of association rules from the dataset. It first retrieves the frequent itemsets from the dataset using the get_frequent_itemsets method. For each frequent itemset, the method generates all possible rules using the get_rules method. These rules are then filtered based on their support and confidence. Only those rules whose support is greater than or equal to the support threshold and whose confidence is greater than or equal to the confidence threshold are included in the final list of valid association rules. This list is then returned. The get_association_rules method provides a set of actionable rules that can be used to understand relationships between different items in the dataset.

```python
#Let's import the RuleMiner object
from rule_miner import RuleMiner
```

```python
#Input support and confidence levels
support_t = 0.09
confidence_t = 0.7

#Initialize RuleMiner with relative support
miner = RuleMiner(support_t, confidence_t,relative_support=True)
```

How did we choose the support and confidence thresholds? Eventually, we chose a support threshold of 9% and a confidence level of 70%. This means we are only interested in itemsets that appear in at least 9% of the total transactions in the dataset. This threshold helps filter out infrequent itemsets that may not provide meaningful insights.

The confidence threshold of 70% means we look for association rules where the likelihood of the right-hand side (rhs) items being present, given that the left-hand side (lhs) items are present, is at least 70%. This ensures the rules we generate are both strong and reliable.

Using 0.09 as the support threshold and 0.7 as the confidence threshold is appropriate and well-justified for several reasons. First, the dataset is very large, so setting a support threshold of 0.1 (10%) or higher excludes too many itemsets, leading to fewer meaningful rules. By reducing the threshold slightly to 0.09 (9%), we strike a balance that captures more frequent itemsets while avoiding rare combinations that are less likely to provide actionable insights.

Second, the chosen thresholds resulted in generating the most rules where income is the consequent. This aligns with the analytical goal of uncovering meaningful relationships that influence income, making the thresholds practical for the specific use case. A higher support threshold would miss out on important but slightly less frequent patterns, whereas a lower support threshold might overburden the process with noise or less actionable itemsets.

Finally, the confidence threshold of 0.7 (70%) ensures that the generated rules are strong and reliable. A lower confidence threshold would include weaker associations that might not be practically useful, while a higher threshold might overly restrict the rules and exclude potentially significant insights.

```python
#Generate association rules
rules = miner.get_association_rules(df_encoded)
print("Generated Rules:", rules)

#Initialize an empty list to store the rule details
rule_list = []

#Loop through each rule and extract the necessary details
for rule in rules:
    antecedent = rule[0]  #Left hand side (antecedent)
    consequent = rule[1]  #Right hand side (consequent)
    
    #Combine antecedent and consequent into a single list of items
    itemset = list(antecedent) + list(consequent)  #Ensure the itemset is a flat list
    
    #Calculate confidence and support
    confidence = miner.get_confidence(df_encoded, rule)
    support = miner.get_support(df_encoded, itemset)
    
    #Store rule details in a list
    rule_list.append({
        "antecedent": antecedent,
        "consequent": consequent,
        "confidence": confidence,
        "support": support
    })

#Convert the list of rules into a DataFrame
rules_df = pd.DataFrame(rule_list)
```
